Data Loading

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("employees.csv")
print(df)
```

Male Employees

```python
male_only = df[df["Gender"]=="Male"]
print(male_only)
```

Female Employees

```python
female_only = df[df["Gender"]=="Female"]
print(female_only)
```

Total number of Male and Female

```python
gender_counts = df["Gender"].value_counts()
print(gender_counts)
plt.figure(figsize=(5, 5))
plt.title("Male vs Female")
plt.pie(gender_counts, labels = gender_counts.index, autopct="%2.2f%%", colors=["skyblue", "green"])
plt.show()
```

People who like Tea or Coffee

```python
tea_or_coffee = df[(df["Drinks"]=="Tea") | (df["Drinks"]=="Coffee")]
print(tea_or_coffee)
```

People who like Beer or Wine

```python
beer_and_wine = df[(df["Drinks"]=="Beer") | (df["Drinks"]=="Wine")]
print(beer_and_wine)
```

Salary greater less than equal to 50K

```python
less_than_50 = df[df["Salary"]<=50000]
print(less_than_50)
```

Salary greater than 50K

```python
greater_than_50 = df[df["Salary"]>50000]
print(greater_than_50)
```

Impact of City on Salary and Expense

```python
gb = df.groupby("City").agg({"Salary":"mean", "Expense":"mean"})
plt.figure(figsize=(5, 5))
plt.title("Impact of City on Salary and Expense")
sns.heatmap(gb, annot=True, fmt=".2f")
plt.show()
```

Correlation between Salary and Expense

```python
correlation = df["Salary"].corr(df["Expense"])

print("Correlation between Salary and Expense: ", correlation)
```

Impact of Drinks on Expense

```python
gb2 = df.groupby("Drinks").agg({"Expense":"mean"})
plt.figure(figsize=(5,5))
plt.title("Impact of Drinks on Expense")
sns.heatmap(gb2, annot=True, fmt=".2f")
plt.show()
```

Age Distribution on Different Hobbies

```python
gb3 = df.groupby("Hobbies").agg({"Age":"mean"})
plt.figure(figsize=(5,5))
plt.title("Age Distribution on different Hobbies")
sns.heatmap(gb3, annot=True, fmt=".2f")
plt.show()
```

Distribution of Drinks

```python
plt.figure(figsize=(5, 5))
plt.title("Distribution of Drinks")
drinks_count = df["Drinks"].value_counts()
plt.pie(drinks_count, labels= drinks_count.index,startangle=360, autopct="%2.2f%%", colors=["skyblue", "red", "#4CAF20", "yellow"])
plt.show()
```

Drinks by Gender

```python
plt.figure(figsize= (6, 5))
sns.countplot(x=df["Drinks"], hue=df["Gender"])
plt.title("Drinks by Gender")
plt.show()
```

Position and Salary Distribution

```python
plt.figure(figsize= (18, 10))
plt.title("Position and Salary Distribution")
x = df["Position"]
y = df["Salary"]
plt.bar(x, y )
plt.show()
```

Physical and Online Workers

```python
plt.figure(figsize= (5, 5))
plt.title("Physical and Online Workers")
working_mode = df["Working Mode"].value_counts()
plt.pie(working_mode, labels = working_mode.index, autopct="%2.2f%%")
plt.show()
```

Laptop used by employees

```python
plt.figure(figsize= (12, 6))
sns.countplot(x=df["Laptop"], hue=df["Position"])
plt.title("Laptop used by employees")
plt.show()
```
